```python
# --- Import Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose


np.random.seed(42)
date_rng = pd.date_range(start="2020-01-01", end="2022-12-31", freq='D')
trend = np.linspace(100, 200, len(date_rng))
seasonality = 10 * np.sin(2 * np.pi * date_rng.dayofyear / 365)
noise = np.random.normal(0, 5, len(date_rng))
price = trend + seasonality + noise

df = pd.DataFrame({"Date": date_rng, "Stock_Price": price})
df.set_index("Date", inplace=True)

# --- Step 2: Plot the Time-Series Data ---
plt.figure(figsize=(10,5))
plt.plot(df["Stock_Price"], color='teal')
plt.title("Stock Price Time Series")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


```

```python

df["Moving_Avg_30D"] = df["Stock_Price"].rolling(window=30).mean()

plt.figure(figsize=(10,5))
plt.plot(df["Stock_Price"], label="Original", alpha=0.5)
plt.plot(df["Moving_Avg_30D"], color='red', label="30-Day Moving Average")
plt.title("Stock Price with 30-Day Moving Average")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
